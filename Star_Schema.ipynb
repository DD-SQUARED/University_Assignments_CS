```python
# Install necessary libraries
!pip install matplotlib networkx

import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add nodes for the Fact Table and Dimension Tables
nodes = {
    'FactOrder': 'FactOrder\n(OrderNum, OrderDate, DeptCode, VendCode, OrderAmount)',
    'DimDate': 'DimDate\n(DateID, OrderDate, Year, Month, Day)',
    'DimDepartment': 'DimDepartment\n(DeptCode, DeptName, DeptContactDeptEmail)',
    'DimVendor': 'DimVendor\n(VendCode, VenName, VenContact, VenEmail)'
}

# Add nodes to the graph
for node, label in nodes.items():
    G.add_node(node, label=label)

# Add edges to connect Fact Table with Dimension Tables
edges = [
    ('FactOrder', 'DimDate'),
    ('FactOrder', 'DimDepartment'),
    ('FactOrder', 'DimVendor')
]

# Add edges to the graph
G.add_edges_from(edges)

# Set node positions
pos = {
    'FactOrder': (0, 0),
    'DimDate': (-2, 1),
    'DimDepartment': (0, 1),
    'DimVendor': (2, 1)
}

# Draw the graph
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, labels=nx.get_node_attributes(G, 'label'),
        node_size=4000, node_color='lightblue', font_size=10, font_weight='bold',
        edge_color='black', linewidths=2, arrows=True, arrowsize=20)

# Show the plot
plt.title('Star Schema Diagram')
plt.show()
```
